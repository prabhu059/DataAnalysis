```python
import sklearn as sk
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
t=[[5],[7],[12],[16],[20]]
m=[[40],[120],[180],[210],[240]]
LR.fit(t,m),
LR.predict([[5.5]])

```

```python
plt.scatter(t,m,color='black')
y_pred=LR.predict([[5.5]])
print(y_pred)
plt.plot(5.5,y_pred,color='blue',linewidth=3,marker="H")
plt.xlabel('time')
plt.ylabel('mass')
plt.title("LinearRegression")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
num_points = 100

# Generate random data points for class A (above the line)
class_A_x = np.random.uniform(0, 10, num_points)
class_A_y = 0.4 * class_A_x + np.random.normal(3, 1, num_points)

# Generate random data points for class B (below the line)
class_B_x = np.random.uniform(0, 10, num_points)
class_B_y = 0.4 * class_B_x + np.random.normal(-2, 1, num_points)

# Plot the data
plt.figure(figsize=(8, 6))

# Plot points for class A (dots)
plt.scatter(class_A_x, class_A_y, color='blue', marker='o', label='Class A')

# Plot points for class B (squares)
plt.scatter(class_B_x, class_B_y, color='red', marker='s', label='Class B')

# Plot the separating line
x_line = np.linspace(0, 10, 100)
y_line = 0.4 * x_line + 1.5
plt.plot(x_line, y_line, color='black', linestyle='--', label='sepereting Line')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Graph with Two Classes')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
